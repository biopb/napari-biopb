```python
import biopb.image as proto
import grpc
import numpy as np
from biopb.image.utils import deserialize_to_numpy, serialize_from_numpy

SERVER="10.84.11.138:8888"
```

```python
def grpc_call(image):
    request = proto.ProcessRequest(
        image_data = proto.ImageData(pixels=serialize_from_numpy(image)),
    )

    with grpc.insecure_channel(target=SERVER) as channel:
        stub = proto.ProcessImageStub(channel)
        response = stub.Run(request)

    # generate label
    label = deserialize_to_numpy(response.image_data.pixels)

    return label

def grpc_call_cellpose(image):
    request = proto.ProcessRequest(
        image_data = proto.ImageData(pixels=serialize_from_numpy(image)),
    )

    with grpc.secure_channel(target="cellpose.biopb.org", credentials=grpc.ssl_channel_credentials()) as channel:
        stub = proto.ProcessImageStub(channel)
        response = stub.Run(request)

    # generate label
    label = deserialize_to_numpy(response.image_data.pixels)

    return label
```

```python
img = np.random.normal(size=[384,384,3])
grpc_call(img)
```
